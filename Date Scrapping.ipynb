```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.service import Service as EdgeService
```

```python
# Set up Edge WebDriver
edge_service = EdgeService("C:/Users/Sarah/Downloads/edgedriver_win64/msedgedriver.exe")
driver = webdriver.Edge(service=edge_service)

# URL of the website
url = 'https://www.whiteflash.com/loose-diamonds/search/'

driver.get(url)

all_items=[]
# Loop through each page
for page_number in range(1, 13): #135
    # Get the URL of the current page
    page_url = f'{url}?page={page_number}'
    driver.get(page_url)
    
    # Scroll down the page to trigger loading of dynamically appearing elements
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for the items to load
    items = WebDriverWait(driver, 5).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'items-grid'))
    )
    
    # Extract data from each item
    for item in items:
        # Get the content inside the item
        all_items.append(item.get_attribute('outerHTML'))
        #print(item.text)# You can modify this to store the data as per your requirement

# Close the WebDriver
driver.quit()
```

```python
# Set up Edge WebDriver
edge_service = EdgeService("C:/Users/Sarah/Downloads/edgedriver_win64/msedgedriver.exe")
driver = webdriver.Edge(service=edge_service)

# URL of the website
url = 'https://www.whiteflash.com/loose-diamonds/search/?carat=0,20&color=D,L&clarity=IF,SI3'

driver.get(url)

all_items=[]
# Loop through each page
for page_number in range(1, 1024): #135
    # Get the URL of the current page
    page_url = f'{url}&page={page_number}'
    driver.get(page_url)
    
    # Scroll down the page to trigger loading of dynamically appearing elements
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for the items to load
    items = WebDriverWait(driver, 5).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'items-grid'))
    )
    
    # Extract data from each item
    for item in items:
        # Get the content inside the item
        all_items.append(item.get_attribute('outerHTML'))
        #print(item.text)# You can modify this to store the data as per your requirement

# Close the WebDriver
driver.quit()
```

```python
all_soup = []
for item_html in all_items:
    soup = BeautifulSoup(item_html, 'html.parser')
    all_soup.append(soup)
```

```python
import numpy as np
data=[]
for index_soup in all_soup:
    for items in index_soup:
        all_info = items.find_all('div', class_='col-xxs-6 col-xs-4 col-sm-4 col-md-3 col-lg-3 item-block-padding')
        for info in all_info:
            hdn = info.find('input', class_='hdnTipData')['value']
            price = info.find('div', class_='s-m-d-p').text.strip(' ').strip('$').replace(',','')
            cut = info.find('div', class_='s-m-d-c').text.split('Round')[1]

            data_dict =  eval(hdn.strip("[]").replace("'",'"'))
            data_dict['price'] = int(price)
            data_dict['cut'] = cut
            data.append(data_dict)
df=pd.DataFrame(data)
        
```
